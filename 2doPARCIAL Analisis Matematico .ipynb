Ejercicio N° 1
Dada la función $y=tan(x)$ y sabiendo que $f(\frac{π}{4})=1$ y $f(\frac{3}{4}π)=−1$ ,¿se puede asegurar por el teorema de Bolzano que existe $x=c$ perteneciente al intervalo $[\frac{π}{4},\frac{3}{4}π]$ tal que $f(c)=0$ ?

```python
import numpy as np
import matplotlib.pyplot as plt

x, y = sp.symbols('x y')
f1 = np.pi/4
f2 = (3/4)*np.pi
x = np.linspace(f1,f2,50)
def f(x):
    y = np.tan(x)
    return y
f(x)
```

```python
plt.plot(x,f(x))
plt.axhline(y=0, color="red")
plt.show()
```

No se cumple el teorema de Bolzano, ya que la función es discontinua... para que se cumpla el teorema de Bolsano deben cumplirse dos condiciones:
    
    
1 - Que la f(x) seea continua en un intervalo cerrado [a,b].
2 - Que sus signos de f(a) y f(b) sean distintos, entonces existe al menos un punto c perteneciente al intervalo abierto (a,b), tal que f(c)=0.

La regla de L'Hôpital nos ayuda a encontrar muchos límites donde la sustitución directa termina en las formas indeterminadas $\frac{0}{0}$, $\frac{0}{\infty}$ o $\frac{∞}{∞}$.

Sean $f$ y $g$ dos funciones continuas definidas en el intervalo [a,b], derivables en (a,b) y sea c perteneciente a (a,b) tal que $f(c)=g(c)=0$ y $g'(x)≠0$ si $x≠c$.

Si existe el límite $L$ de $\frac{f'}{g'}$ en c, entonces existe el límite de $\frac{f}{g} $ (en c) y es igual a $L$.

Por lo tanto,

$$ \lim _{x\to c} \frac{f(x)}{g(x)}=\lim _{x\to c}\frac{f'(x)}{g'(x)}=L  $$

Aplicar la regla de L'Hôpital para resolver la indeterminación de:
